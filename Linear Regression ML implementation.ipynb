```python
## dividing our features into indepedent and dependent features 
```

```python
dataset.head()
```

```python
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
x.head()
```

```python
y
```

```python
##train test split 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42) 
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
## Standardizing the data set 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
##Model training of the data set 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
## print the coefficients and intercept 
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
##on which parameters the model has been trained 
regression.get_params()
```

```python
## Prediction with test data 
reg_pred=regression.predict(x_test)
```

```python
reg_pred
plt.show()
```

```python
##plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred)
plt.show()
```

```python
##residuals - it means showing errors 
residuals=y_test-reg_pred
```

```python
residuals
```

```python
##Plot the residuals
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.displot(residuals,kind="kde")
plt.show()
```

```python
##Scattering a plot with respect to prediction and residuals 
## uniform distribution 
plt.scatter(reg_pred , residuals)
```

```python
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
## R square and adjusted RSquare
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```
